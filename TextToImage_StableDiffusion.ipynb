```python
# Import libraries for Stable Diffusion
from diffusers import StableDiffusionPipeline
import torch

# Set the Stable Diffusion model ID and data type (float16 for potential memory efficiency)
model_id = "runwayml/stable-diffusion-v1-5"
sd_pipeline = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)

# Try to move the pipeline to GPU for faster processing
ad_pipeline = sd_pipeline.to("cuda")
```

```python
# Negative prompts (what we do not need in the output images)
def get_completion_sd(prompt):
  negative_prompt = """
      low quality, blurry, unrealistic anatomy, nonsensical objects,
      simple background, excessive noise, text, watermarks, signatures,
      duplicate objects, extra limbs, inconsistent lighting
    """

  return sd_pipeline(prompt, negative_prompt=negative_prompt).images[0]
```

```python
# Get user input for the prompt
prompt = input("Enter your desired prompt (text describing the image): ")

# Example prompts
# prompt = "dog wearing a police jacket"
# prompt = "a mountain view"

# print(prompt)
sd_image = get_completion_sd(prompt)
sd_image.save("./soham.jpg")
```
